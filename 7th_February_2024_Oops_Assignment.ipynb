```python
# 19. Explain the use of constructor chaining in Python. Provide a practical example.
# Ans :- Constructor chaining in Python refers to the process of one constructor calling another constructor in the same class or in its parent class. This allows you to reuse initialization logic from one constructor in another, reducing code duplication and promoting maintainability.
#        The super() function is commonly used for constructor chaining, especially in the context of inheritance.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, student_id):
        # Call the constructor of the parent class using super()
        super().__init__(name, age)

        # Additional initialization specific to the Student class
        self.student_id = student_id

    def display_student_info(self):
        # Reuse display_info method from the parent class
        self.display_info()
        print(f"Student ID: {self.student_id}")

# Creating an instance of the Student class
student = Student("Alice", 20, "S12345")

# Displaying information using the method from the Student class
student.display_student_info()
```

```python
# 2. Differentiate between single inheritance and multiple inheritance in Python. Provide examples for each.
# Ans:- 1. Single Inheritance:- In single inheritance, a class can inherit from only one superclass. The drived class inherit the attibutes
#          and methods of a single base class. This is a simpler and more straightforward from of inheritance.
# Example :- 
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
my_dog = Dog()

# Accessing methods from both Animal and Dog classes
my_dog.speak()  # Output: Animal speaks
my_dog.bark()   # Output: Dog barks

# 2. Multiple Inheritance :- A class can inherit from more than one superclass. The drived class inherit attributes and methods from
#                            multiple base classes. While this allows for grater flexibility in code reuse, it can also introduce complexities,
# such as dealing with conflicts when two base classes have methods with the same name.
# Example:-
class Animal:
    def speak(self):
        print("Animal speaks")

class Pet:
    def play(self):
        print("Pet plays")

class Dog(Animal, Pet):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
my_dog = Dog()

# Accessing methods from Animal, Pet, and Dog classes
my_dog.speak()  # Output: Animal speaks
my_dog.play()   # Output: Pet plays
my_dog.bark()   # Output: Dog barks
```

```python
# 6. Discuss the use of the `super()` function in Python inheritance. When and why is it used? Provide an example
# Ans :- The super() function in Python is used in the context of inheritance to refer to the parent class. It provides a way to call methods and access attributes of the superclass (parent class) from within the subclass (child class). 
#.      The primary purpose of super() is to facilitate cooperative multiple inheritance and ensure that the correct methods are called in the correct order in the class hierarchy

class Parent:
    def __init__(self, name):
        self.name = name

    def display(self):
        print(f"Parent's name: {self.name}")

class Child(Parent):
    def __init__(self, name, child_name):
        super().__init__(name)  # Calling the constructor of the Parent class
        self.child_name = child_name

    def display(self):
        super().display()  # Calling the display method of the Parent class
        print(f"Child's name: {self.child_name}")

# Example usage
child_obj = Child(name='John', child_name='Alice')
child_obj.display()
```

```python
# 11.  Create a Python class called `Shape` with a method `area()` that calculates the area of a shape. Then, create child classes `Circle` and `Rectangle` that inherit from `Shape` and implement the `area()` metho
# provide a example.

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example of using the classes
circle = Circle(radius=5)
rectangle = Rectangle(length=4, width=6)

# Accessing the area method of concrete classes
print(f"Area of Circle: {circle.area():.2f}")       # Output: 78.50
print(f"Area of Rectangle: {rectangle.area()}")      # Output: 24
```

```python
# 12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an
#  example using the `abc` module

# Ans - In python provide a way to define abstract classes and abstract methods. An abstract class connot be instatiated, and it is meant to bre
#       subclassed by concrete classes. Abstract methods declared in an abstract class must be implemented by any concrete
#       subclass. ABCs help enforce a certain structure in class hierarchies and provide a form fo documentation about the expected interface of derived classes.
# The abc module in Pyhton is used to work with abstract base classes.
#. The "ABC' classes from this module is used as a metaclass for creating abstract base classes. Abstract methods are defined using the '@abstractmethod' decorator.

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

# Creating instances of concrete classes
square = Square(side=4)
circle = Circle(radius=3)

# Accessing the area method of concrete classes
print(f"Area of Square: {square.area()}")  # Output: 16
print(f"Area of Circle: {circle.area()}")  # Output: 28.26
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self):
        self.books = {}

    @abstractmethod
    def add_book(self, book_id, title, author):
        pass

    @abstractmethod
    def borrow_book(self, book_id, user_id):
        pass

    @abstractmethod
    def return_book(self, book_id, user_id):
        pass

class SimpleLibrary(LibrarySystem):
    def add_book(self, book_id, title, author):
        if book_id not in self.books:
            self.books[book_id] = {'title': title, 'author': author, 'available': True}
            print(f"Added book '{title}' by {author} with ID {book_id}")
        else:
            print("Book ID already exists.")

    def borrow_book(self, book_id, user_id):
        if book_id in self.books:
            if self.books[book_id]['available']:
                self.books[book_id]['available'] = False
                print(f"Book '{self.books[book_id]['title']}' borrowed by user {user_id}")
            else:
                print("Book is already borrowed.")
        else:
            print("Book ID not found.")

    def return_book(self, book_id, user_id):
        if book_id in self.books:
            if not self.books[book_id]['available']:
                self.books[book_id]['available'] = True
                print(f"Book '{self.books[book_id]['title']}' returned by user {user_id}")
            else:
                print("Book is already available.")
        else:
            print("Book ID not found.")

# Usage
library = SimpleLibrary()
library.add_book(1, "Python Programming", "John Smith")
library.add_book(2, "Data Structures", "Alice Johnson")

library.borrow_book(1, "A123")
library.borrow_book(1, "B456")
library.return_book(1, "A123")
library.borrow_book(1, "B456")
```

# 17.  Describe the concept of "aggregation" in composition and how it differs from simple composition.
Ans :- In object-oriented programming, "aggregation" is a form of composition where one object (the whole) is composed of multiple other objects (the parts), but the parts can exist independently of the whole. Aggregation represents a "has-a" relationship between objects, where one object contains or is composed of other objects, but the lifetime of the contained objects is not tied to the lifetime of the containing object.

Here's how aggregation differs from simple composition:

1. **Lifetime Dependency**:
   - In simple composition, the lifetime of the contained objects is typically tied to the lifetime of the containing object. If the containing object is destroyed, the contained objects are also destroyed.
   - In aggregation, the contained objects can exist independently of the containing object. They are not destroyed when the containing object is destroyed, and they can be shared by multiple containing objects.

2. **Ownership**:
   - In simple composition, the containing object typically owns the contained objects. It is responsible for creating, managing, and destroying the contained objects.
   - In aggregation, the containing object may or may not own the contained objects. It simply references or holds a reference to the contained objects, but it does not control their creation or destruction.

3. **Multiplicity**:
   - Simple composition implies a one-to-one or one-to-many relationship between the containing object and the contained objects. Each instance of the containing object is associated with one or more instances of the contained objects.
   - Aggregation implies a one-to-many or many-to-many relationship between the containing object and the contained objects. Multiple instances of the containing object can share or reference the same instances of the contained objects.

4. **Navigation**:
   - In simple composition, the containing object typically directly owns or contains the instances of the contained objects. It has direct access to the contained objects' attributes and methods.
   - In aggregation, the containing object holds a reference to the instances of the contained objects but may not directly own or manage them. It may need to navigate through other objects or accessors to interact with the contained objects.

5. **Flexibility**:
   - Aggregation offers more flexibility compared to simple composition. Since the contained objects can exist independently, they can be easily shared or reused across multiple containing objects.
   - Aggregation allows for more dynamic relationships between objects, making it easier to model complex systems and promote code reusability.

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Furniture: {self.name}")


class Appliance:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Appliance: {self.name}")


class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display_info(self):
        print(f"Room: {self.name}")
        if self.furniture:
            print("Furniture:")
            for item in self.furniture:
                item.display_info()
        if self.appliances:
            print("Appliances:")
            for item in self.appliances:
                item.display_info()


class House:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def display_info(self):
        print("House:")
        if self.rooms:
            for room in self.rooms:
                room.display_info()


# Example usage:
living_room = Room("Living Room")
living_room.add_furniture(Furniture("Sofa"))
living_room.add_furniture(Furniture("Coffee Table"))
living_room.add_appliance(Appliance("Television"))

kitchen = Room("Kitchen")
kitchen.add_furniture(Furniture("Dining Table"))
kitchen.add_appliance(Appliance("Refrigerator"))
kitchen.add_appliance(Appliance("Oven"))

my_house = House()
my_house.add_room(living_room)
my_house.add_room(kitchen)

my_house.display_info()
```

Here's how each of these techniques enables dynamic replacement or modification of composed objects:

1. Dependency Injection:

Dependency injection is a technique where dependencies of an object are provided from the outside rather than created internally.
By injecting dependencies as parameters or through setters, you can easily replace or modify the composed objects at runtime without modifying the class implementation.
This allows for flexibility and loose coupling between the components, as the class does not depend on concrete implementations of its dependencies.

2. Factory Patterns:

Factory patterns, such as the abstract factory pattern or the factory method pattern, provide an interface for creating families of related or dependent objects without specifying their concrete classes.
By using factories to create instances of composed objects, you can dynamically choose which concrete implementations to use at runtime.
This allows for flexibility in selecting and replacing components without modifying the client code that uses the factory.


```python
# 20.Create a Python class for a social media application, using composition to represent users, posts, and comments.

class Comment:
    def __init__(self, author, content):
        self.author = author
        self.content = content

    def display_info(self):
        print(f"Comment by {self.author}: {self.content}")


class Post:
    def __init__(self, author, content):
        self.author = author
        self.content = content
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_info(self):
        print(f"Post by {self.author}: {self.content}")
        if self.comments:
            print("Comments:")
            for comment in self.comments:
                comment.display_info()


class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def create_post(self, content):
        post = Post(self.username, content)
        self.posts.append(post)
        return post

    def display_info(self):
        print(f"User: {self.username}")
        if self.posts:
            print("Posts:")
            for post in self.posts:
                post.display_info()


class SocialMediaApp:
    def __init__(self):
        self.users = []

    def create_user(self, username):
        user = User(username)
        self.users.append(user)
        return user

    def display_info(self):
        print("Social Media App:")
        if self.users:
            for user in self.users:
                user.display_info()


# Example usage:
app = SocialMediaApp()

user1 = app.create_user("Alice")
user2 = app.create_user("Bob")

post1 = user1.create_post("Hello, world!")
post2 = user2.create_post("Goodbye, world!")

comment1 = Comment("Charlie", "Nice post!")
comment2 = Comment("David", "I agree!")

post1.add_comment(comment1)
post2.add_comment(comment2)

app.display_info()
```
